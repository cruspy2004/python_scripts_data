```python
import matplotlib.pyplot as plt

# Group data: total sales per category
category_sales = df.groupby("Category")["Sales"].sum()

# Plot
plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values, color="skyblue")

# Labels + title
plt.title("Sales per Category (Matplotlib)")
plt.xlabel("Category")
plt.ylabel("Total Sales")

# Add values on top of bars
for i, val in enumerate(category_sales.values):
    plt.text(i, val + 2000, round(val, 2), ha='center')

plt.show()
```

```python
# Group data: sales by quarter
quarterly_sales = df.groupby(df["Order Date"].dt.to_period("Q"))["Sales"].sum()

# Plot
plt.figure(figsize=(15,6))
plt.plot(quarterly_sales.index.astype(str), quarterly_sales.values, marker="o", linestyle="-", color="green", label="Sales")

# Labels + title + legend
plt.title("Quarterly Sales Trend (Matplotlib)")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(True)

# Add values on points
for i, val in enumerate(quarterly_sales.values):
    plt.text(i, val + 2000, round(val, 2), ha='center')

plt.show()
```

```python
import seaborn as sns   # import seaborn for prettier plots
import matplotlib.pyplot as plt   


category_sales = df.groupby("Category")["Sales"].sum().reset_index()


plt.figure(figsize=(8,5))  
sns.barplot(
    x="Category",           
    y="Sales",             
    data=category_sales,  
    palette="pastel"        
)

plt.title("Sales per Category (Seaborn)")   
plt.xlabel("Category")   
plt.ylabel("Total Sales")   

# Add values above each bar
for i, val in enumerate(category_sales["Sales"].values):
    plt.text(i, val + 2000, round(val, 2), ha='center')

plt.show()   # display the chart
```

```python
quarterly_sales = df.groupby(df["Order Date"].dt.to_period("Q"))["Sales"].sum().reset_index()
# reset_index()  turn the quarter info into a column so seaborn can read it


quarterly_sales["Order Date"] = quarterly_sales["Order Date"].astype(str)

plt.figure(figsize=(15,6))   # bigger figure 15x6 inches
sns.lineplot(
    x="Order Date",          # x-axis  quarter
    y="Sales",               # y-axis  sales
    data=quarterly_sales,    
    marker="o",              
    color="red"              
)

plt.title("Quarterly Sales Trend (Seaborn)")   
plt.xlabel("Quarter")   
plt.ylabel("Total Sales")   

# Add values above each point
for i, val in enumerate(quarterly_sales["Sales"].values):
    plt.text(i, val + 2000, round(val, 2), ha='center')

plt.show()   # display the chart
```

```python

orders_segment = df.groupby("Segment")["Order ID"].count()
# "Segment" = (Consumer, Corporate, Home Office)
# We count how many Order IDs fall into each segment

plt.figure(figsize=(7,7))   # make square figure for pie chart
plt.pie(
    orders_segment.values,             # the actual numbers (order counts)
    labels=orders_segment.index,       # labels = segment names
    autopct="%1.1f%%",                 # show percentages with 1 decimal place
    startangle=90                      # rotate pie so first wedge starts at 12 oclock
)

plt.title("Number of Orders by Segment")   # add chart title
plt.legend(title="Segments")               # add a legend with title
plt.show()   # display chart
```

```python
profit_region = df.groupby("Region")["Profit"].sum()
# "Region" = (Central, East, South, West)
# We sum all profit values per region

plt.figure(figsize=(7,7))   # square figure 
plt.pie(
    profit_region.values,              # the profit numbers
    labels=profit_region.index,        # region names
    autopct="%1.1f%%",                 # display percentages
    startangle=90                      # start at 12 oclock
)

plt.title("Profit by Region")          
plt.legend(title="Regions")            
plt.show()   
```
